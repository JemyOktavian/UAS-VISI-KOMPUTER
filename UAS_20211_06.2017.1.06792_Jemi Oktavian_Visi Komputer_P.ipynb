```python
import mediapipe as mp
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
cap = cv2.VideoCapture(0)
#inisiasi model holistic
with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        #warna feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #membuat deteksi
        results = holistic.process(image)
        print(results.pose_landmarks)
        
        #gambar ke rgb
        mp_drawing.draw_landmarks(image, results.face_landmarks,mp_holistic.POSE_CONNECTIONS)
        
        cv2.imshow('Landmarks',image)
        
        if cv2.waitKey(10) & 0xFF == ord('w'):
            break

            
cap.release()
cv2.destroyAllWindows()
            
```

```python
import cv2
 
#Membaca gambar
img = cv2.imread('Wajah.png')
#Menampilkan gambar
cv2.imshow('Wajah',img)
#Menyimpan gambar
cv2.imwrite('save_Wajah.png',img)
 
#Menunda windows terdestroy
cv2.waitKey(0)
#Mendestroy windows
cv2.destroyAllWindows()
```

```python

def get_pixel(img, center, x, y):

	new_value = 0
	
	try:

		#Diubah ke1
        
		if img[x][y] >= center:
			new_value = 1
			
	except:

		pass
	
	return new_value

#Fungsi Penghitung LBP
def lbp_calculated_pixel(img, x, y):

	center = img[x][y]

	val_ar = []
	
	# top_left
	val_ar.append(get_pixel(img, center, x-1, y-1))
	
	# top
	val_ar.append(get_pixel(img, center, x-1, y))
	
	# top_right
	val_ar.append(get_pixel(img, center, x-1, y + 1))
	
	# right
	val_ar.append(get_pixel(img, center, x, y + 1))
	
	# bottom_right
	val_ar.append(get_pixel(img, center, x + 1, y + 1))
	
	# bottom
	val_ar.append(get_pixel(img, center, x + 1, y))
	
	# bottom_left
	val_ar.append(get_pixel(img, center, x + 1, y-1))
	
	# left
	val_ar.append(get_pixel(img, center, x, y-1))
	
	# Now, we need to convert binary
	# values to decimal
	power_val = [1, 2, 4, 8, 16, 32, 64, 128]

	val = 0
	
	for i in range(len(val_ar)):
		val += val_ar[i] * power_val[i]
		
	return val

path = 'Wajah.png'
img_bgr = cv2.imread(path, 1)

height, width, _ = img_bgr.shape

#Convert RGB
img_gray = cv2.cvtColor(img_bgr,
                            cv2.COLOR_BGR2GRAY)


#RGB 
img_lbp = np.zeros((height, width),
                np.uint8)

for i in range(0, height):
    for j in range(0, width):
        img_lbp[i, j] = lbp_calculated_pixel(img_gray, i, j)

plt.imshow(img_bgr)
plt.show()

plt.imshow(img_lbp, cmap ="gray")
plt.show()

print("LBP Selesai")
```
